```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder  # for Encoding Ordinal(input) Data Columns
from sklearn.preprocessing import OneHotEncoder   # for Encoding Nominal Data Columns
from sklearn.preprocessing import LabelEncoder     # for Encoding Ordinal(output) Column

from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv('encoded_dataset.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train , y_test =train_test_split(df.iloc[:,[2,3,4]] , df.iloc[: ,-1 ]  , test_size=0.1)
```

```python
clt=ColumnTransformer([
    ('trns1', OrdinalEncoder(categories=[['HS','UG','PG']]),['Education']),
     ('trns2',OneHotEncoder(sparse_output=False , drop='first') ,['Gender','City'])
     ])
```

```python
x_train_scaled= clt.fit_transform(x_train)
```
