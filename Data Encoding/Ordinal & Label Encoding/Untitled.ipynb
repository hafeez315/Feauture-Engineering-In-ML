```python
import pandas as pd 
```

```python
df=pd.read_csv('customer_dataset.csv')
```

```python
df
```

```python
x=df.iloc[:,2:]
```

### these are oridinal catagorical columns (input columns ) so we will use ordinal Encoding;

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x.iloc[:, 0:2] , x.iloc[:,-1] , test_size=0.1)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Good','Average', 'Poor'],['HS','UG','PG']])
```

```python
oe.fit(x_train)
```

```python
encoded_x_train=oe.transform(x_train)
encoded_x_test= oe.transform(x_test)
```

```python
encoded_x_train
```

```python
x_train=pd.DataFrame(encoded_x_train,columns=x_train.columns)
x_test=pd.DataFrame(encoded_x_test,columns=x_test.columns)
```

```python
x_train
```

```python
x_test
```
